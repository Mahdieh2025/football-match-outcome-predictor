# Module 5 – Evaluation

## Overview

The objective of this module is to rigorously evaluate the performance of the trained classification models and assess their strengths and limitations.  
Given the multiclass nature of the problem (Home win, Draw, Away win), evaluation focuses on metrics that go beyond simple accuracy and provide a balanced view of model behavior across all classes.

---

## Evaluation Metrics

The following metrics were used throughout the evaluation phase:

- **Accuracy**: Overall proportion of correctly classified matches.
- **Precision, Recall, F1-score (per class)**: To analyze performance for each match outcome.
- **Macro-averaged F1-score**: Chosen as the primary metric to ensure equal importance of all classes, especially the minority class (Draw).
- **Confusion Matrix**: To visualize class-wise prediction errors.

These metrics are well-suited for imbalanced multiclass classification problems such as football match outcome prediction.

---

## Test Set Evaluation

The final evaluation was performed on a held-out test set (20% of the data), not used during training or cross-validation.

Key observations:
- Home wins (H) are predicted with the highest precision and recall.
- Away wins (A) show moderate predictive performance.
- Draws (D) are consistently the hardest class to predict, with lower recall across all models.

This behavior is consistent with domain knowledge, as draws are inherently more stochastic and less frequent.

---

## Confusion Matrix Analysis

The confusion matrices reveal systematic patterns:

- Many draw matches are misclassified as home wins.
- Misclassifications between home and away wins are less frequent.
- The models tend to favor the majority classes.

This indicates that the models are learning meaningful signals but are limited by the available feature set.

---

## Cross-Validation Consistency

Cross-validation results closely match test-set performance, indicating:

- No significant overfitting
- Stable model behavior across different data splits
- Reliable generalization performance

This confirms the robustness of the modeling pipeline.

---

## Model Comparison

Across all evaluation metrics:

- Logistic Regression provides the best balance between performance and interpretability.
- Tree-based models (Random Forest, XGBoost) offer flexibility but do not significantly outperform the baseline given the current features.
- Increasing model complexity alone does not solve the draw-prediction challenge.

---

## Limitations

The main limitations identified are:

- Limited feature richness (no rolling statistics or team strength indicators)
- No external contextual variables (odds, injuries, match importance)
- Inherent uncertainty in football match outcomes

These limitations explain the observed performance ceiling.

---

## Conclusion

The evaluation confirms that the developed models are statistically sound, stable, and correctly implemented.  
While predictive performance is moderate, the evaluation framework is rigorous and aligns with best practices in applied data science.

The results provide a clear roadmap for future improvements through enhanced feature engineering rather than more complex algorithms.
